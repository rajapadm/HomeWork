```python
from splinter import Browser
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
browser_path = {'executable_path': 'chromedriver'}
browser = Browser('chrome', **browser_path)
```

```python
browser.visit('https://mars.nasa.gov/news/')
html=browser.html
soup = bs(html, 'html.parser')
print(soup.prettify())
```

```python
latest_news=soup.find('div',class_='image_and_description_container')
news_title=latest_news.find('a').get_text()
news_desc=latest_news.find('h3').get_text()
print(news_title)
print(news_desc)
```

```python
browser.visit('https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars')
html=browser.html
soup = bs(html, 'html.parser')
print(soup.prettify())
```

```python
img_grid=soup.find('ul',class_='articles').find_all('li', class_='slide')
print(img_grid)
```

```python
img_url=[]

for href in img_grid:
    url=href.find('a')
    img_url.append('https://www.jpl.nasa.gov'+url['data-fancybox-href'])
    
img_url
```

```python
browser.visit('https://twitter.com/marswxreport?lang=en')
html=browser.html
soup = bs(html, 'html.parser')
mars_weather=soup.find('div', attrs={"class": "tweet", "data-name": "Mars Weather"}).find('p', 'tweet-text').get_text()
print(mars_weather)
```

```python
url='https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
browser.visit(url)
items = len(browser.find_by_css("div.description a.itemLink"))

hemisphere_image_urls = []

for i in range(items):
    hemisphere = {}
    browser.find_by_css("div.description a.itemLink")[i].click()
    sample = browser.find_link_by_text('Sample').first
    hemisphere['title'] = browser.find_by_css("h2.title").text
    hemisphere['img_url'] = sample['href']
    hemisphere_image_urls.append(hemisphere)
    browser.back()

hemisphere_image_urls    
```
